## 무신사 이미지 크롤링

```python
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import pandas as pd
from tqdm import tqdm
from selenium.webdriver.common.keys import Keys
import urllib.request
from selenium.common.exceptions import NoSuchElementException
```

```python
driver = webdriver.Chrome('chromedriver.exe')
driver.implicitly_wait(3)
driver.get("https://www.musinsa.com/app/styles/lists") #무신사 링크
driver.find_element_by_xpath('/html/body/div[3]/div[3]/div[1]/button[3]').click()

image_list = []
image_name_list = []

address = r"D:\무신사 크롤링 이미지\golf"

# 스타일별
for style in range(1,9): # (range(1,9))
    driver.find_element_by_css_selector(f'#catelist > div:nth-child(3) > div > dl > dd > ul > li:nth-child({style}) > a > span').click()
    
    #페이지별
    for page in tqdm(range(1,11)): #(range(1,11))
        if (page != 1):
            driver.find_element_by_xpath(f'/html/body/div[3]/div[3]/form/div[4]/div/div[1]/div/div/a[{page+2}]').send_keys(Keys.ENTER)
                    
        # 이미지 리스트별
        for image_idx in tqdm(range(1,61)): # range(1,61)
                        
            # 이미지 클릭해서 하나씩 가져오기
            try : 
                # 오류나면 화면 크게 해줘야함!
                driver.find_element_by_css_selector(f"body > div.wrap > div.right_area > form > div.right_contents.hover_box > div > ul > li:nth-child({image_idx}) > div.style-list-item__thumbnail > a > div > img").click()
                time.sleep(0.5)
                
                imgurl = driver.find_element_by_css_selector('#style_info > div.detail_slider_wrap > div.detail_slider.slick-initialized.slick-slider > div > div > img.detail_img.slick-slide.slick-current.slick-active').get_attribute("src")
                image_list.append(imgurl)

                image_name = imgurl.split('/')[-1]
                image_name_list.append(image_name)            

                urllib.request.urlretrieve(imgurl,address+"/"+str(image_name)) # download images in address folder
                
                # 이전페이지로 돌아가기
                driver.back()
                time.sleep(0.5)
            
            # 비디오 예외
            except NoSuchElementException:
                
                driver.back()
```
