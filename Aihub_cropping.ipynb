- total_label.txt에 있는 이미지 파일 경로에 있는 이미지를 cropping

- cropping 좌표는 이미지 경로 뒤에 있는 ,로 구분된 4개의 실수 좌표 그 뒤에는 클래스
- 이미지를 cropping하고 별도의 폴더에 이미지 파일을 저장.
- 하나의 이미지 경로에 두개의 좌표,클래스 묶음이 존재하기도 함.
- bounding box의 좌표는 좌측 X,Y좌표/ 우측 X,Y좌표 순서로 4개로 구성.
- 각 cropping된 이미지 좌표, 이미지 이름, 상하의 여부, 상하의 세부 카테고리, 컬러, 소재, 핏 (카테고리값이 리스트로 되어있으면 첫번째 값만 추출)
-  x_min,x_max,y_min,y_max,class_id로 좌표가 구성되어 있음.
-  PIL에서 Cropping하기 위해서는 x_min, y_min, x_max, y_max로 바꿔줘야 함. 
- 지금 : x좌표, y좌표, 가로, 세로 -> x,y 좌표의 의미 : 중심이 아닌 가장 왼쪽 시작 좌표를 의미

```python
import os
from PIL import Image, ImageDraw
import json
import numpy as np
import pandas as pd


###
save_img_dir = r"D:\cropped_image\casual"
dir_list=os.listdir(r'D:\image_list')
json_dir=r'D:\label_json'
feature_aihub = pd.DataFrame(columns=['파일경로','파일명','상하의','스타일','색상','소재','핏'])

for file in dir_list:
    read_form = f'D:\\image_list\\{file}' ###
    f = open(read_form,mode='r',encoding='utf-8')
    line = None
    for i in tqdm(range(len(f.readlines())):
        line=f.readline()
        line_list=line.split(' ')
        img_path = line_list[0][:-1]
        #file_name > 1181022.jpg
        file_name = img_path.split('/')[-1]
        #print(file_name)
        #file_num > 1181022 (str)
        file_num = file_name[:-4]
        if file_num == '' or file_num == ' ': continue
        #print(file_num)
        json_path = f'D:\\label_json\\{file_num}.json'

        with open(json_path,'r',encoding='utf-8') as j:
            contents = j.read()
            json_data=json.loads(contents)
            #print(json_data)
    

            for cate,value in json_data['데이터셋 정보']['데이터셋 상세설명']['렉트좌표'].items(): 
            
                if ((cate == "하의") & (len(value[0]) != 0)):
                    xmin = value[0]['X좌표']
                    xmax = value[0]['X좌표']+value[0]['가로']
                    ymin = value[0]['Y좌표']
                    ymax = value[0]['Y좌표']+value[0]['세로']
                    local_img_path = os.path.join('D:\\image_data\\casual',file_name)
                    #img = Image.open(img_path)
                    img = Image.open(local_img_path)
                    cropped_img = img.crop((xmin, ymin, xmax, ymax))
                    #display(cropped_img)
                    save_img_path = os.path.join(save_img_dir,f"{file_num}_1.jpg")
                    cropped_img.save(save_img_path)
            
                    try:
                        if isinstance(json_data['데이터셋 정보']['데이터셋 상세설명']['라벨링']['하의'][0]['카테고리'],list):
                            style = json_data['데이터셋 정보']['데이터셋 상세설명']['라벨링']['하의'][0]['카테고리'][0]
                        else:
                            style = json_data['데이터셋 정보']['데이터셋 상세설명']['라벨링']['하의'][0]['카테고리']
                    except:
                        style = 0
              
                    try:
                        if isinstance(json_data['데이터셋 정보']['데이터셋 상세설명']['라벨링']['하의'][0]['색상'],list):
                            color = json_data['데이터셋 정보']['데이터셋 상세설명']['라벨링']['하의'][0]['색상'][0]
                        else:
                            color = json_data['데이터셋 정보']['데이터셋 상세설명']['라벨링']['하의'][0]['색상']
                    except:
                        color = 0
          
                    try:
                        if isinstance(json_data['데이터셋 정보']['데이터셋 상세설명']['라벨링']['하의'][0]['소재'],list):
                            material = json_data['데이터셋 정보']['데이터셋 상세설명']['라벨링']['하의'][0]['소재'][0]
                        else:
                            material = json_data['데이터셋 정보']['데이터셋 상세설명']['라벨링']['하의'][0]['소재']
                    except:
                        material =0

                    try:
                        if isinstance(json_data['데이터셋 정보']['데이터셋 상세설명']['라벨링']['하의'][0]['핏'],list):
                            fit = json_data['데이터셋 정보']['데이터셋 상세설명']['라벨링']['하의'][0]['핏'][0]
                        else:
                            fit = json_data['데이터셋 정보']['데이터셋 상세설명']['라벨링']['하의'][0]['핏']
                    except:
                        fit = 0
                  
                    colab_path=os.path.join('/content/gdrive/MyDrive/Colab Notebooks/딥러닝 term project/cropped_image',f'{file_name}_1')
                    feature_aihub.loc[f'{file_name}_1',:]=[str(colab_path),file_num[:-4]+'_1'+file_num[-4:],"하의",style,color,material,fit]
              
                elif ((cate == "상의") & (len(value[0]) != 0)):
                    xmin = value[0]['X좌표']
                    xmax = value[0]['X좌표']+value[0]['가로']
                    ymin = value[0]['Y좌표']
                    ymax = value[0]['Y좌표']+value[0]['세로']
                    local_img_path = os.path.join('D:\\image_data\\casual',file_name)
                    #img = Image.open(img_path)
                    img = Image.open(local_img_path)
                    cropped_img = img.crop((xmin, ymin, xmax, ymax))
                    #display(cropped_img)
                    save_img_path = os.path.join(save_img_dir,f"{file_num}_2.jpg")
                    cropped_img.save(save_img_path)
        
                    try:
                        if isinstance(json_data['데이터셋 정보']['데이터셋 상세설명']['라벨링']['상의'][0]['카테고리'],list):
                            style = json_data['데이터셋 정보']['데이터셋 상세설명']['라벨링']['상의'][0]['카테고리'][0]
                        else:
                            style = json_data['데이터셋 정보']['데이터셋 상세설명']['라벨링']['상의'][0]['카테고리']
                    except:
                        style = 0

                    try:
                        if isinstance(json_data['데이터셋 정보']['데이터셋 상세설명']['라벨링']['상의'][0]['색상'],list):
                            color = json_data['데이터셋 정보']['데이터셋 상세설명']['라벨링']['상의'][0]['색상'][0]
                        else:
                            color = json_data['데이터셋 정보']['데이터셋 상세설명']['라벨링']['상의'][0]['색상']
                    except:
                        color= 0

                    try:
                        if isinstance(json_data['데이터셋 정보']['데이터셋 상세설명']['라벨링']['상의'][0]['소재'],list):
                            material = json_data['데이터셋 정보']['데이터셋 상세설명']['라벨링']['상의'][0]['소재'][0]
                        else:
                            material = json_data['데이터셋 정보']['데이터셋 상세설명']['라벨링']['상의'][0]['소재']
                    except:
                        material=0

                    try:
                        if isinstance(json_data['데이터셋 정보']['데이터셋 상세설명']['라벨링']['상의'][0]['핏'],list):
                            fit = json_data['데이터셋 정보']['데이터셋 상세설명']['라벨링']['상의'][0]['핏'][0]
                        else:
                            fit = json_data['데이터셋 정보']['데이터셋 상세설명']['라벨링']['상의'][0]['핏']
                    except:
                        fit = 0
                    
                    colab_path=os.path.join('/content/gdrive/MyDrive/Colab Notebooks/딥러닝 term project/cropped_image',f'{file_name}_1')
                    feature_aihub.loc[f'{file_name}_2',:]=[str(colab_path),file_num[:-4]+'_2'+file_num[-4:],"상의",style,color,material,fit]
                
        j.close()
    f.close()
```
