```python
# 폴더 내 파일 리스트 추출해서 코드 자동화하기

import os
from tqdm import tqdm

directory = 'D:\K-Fashion 이미지\Validation\라벨링데이터\페미닌' ###
file_list = os.listdir(directory)
```

```python
# x_min,y_min,x_max,y_max,class_id
# 지금 : x좌표, y좌표, 가로, 세로 -> x,y 좌표의 의미 : 중심이 아닌 가장 왼쪽 시작 좌표를 의미

import json

for file in tqdm(file_list):
    read_form = f'D:\K-Fashion 이미지\Validation\라벨링데이터\페미닌\\{file}' ###
    
    with open(read_form,'r',encoding='utf8') as f:
        contents=f.read() # string 타입
        json_data=json.loads(contents)
        
    # 하의 : 0, 상의 : 1

    cnt = 0
    info =''

    # 이미지 경로 지정  
    path = '/'.join(["/content/gdrive/MyDrive/deep_learning_team_project/data/training/image/feminine",
                     str(json_data['데이터셋 정보']['파일 번호'])]) ###
    path += '.jpg'

    for cate,value in json_data['데이터셋 정보']['데이터셋 상세설명']['렉트좌표'].items():         
   
        # category에 대한 값이 존재하고, 카테고리 정보가 처음일때
        if ((len(value[0]) != 0) & (cnt==0)):
            
            x_min = str(value[0]['X좌표'])
            x_max = str(value[0]['X좌표']+value[0]['가로'])
            y_min = str(value[0]['Y좌표'])
            y_max = str(value[0]['Y좌표']+value[0]['세로'])
            
            if (cate=='하의'):
                info += ','.join([x_min,y_min,x_max,y_max])
                info = ','.join([info,'0'])
                cnt += 1
                
            elif (cate=='상의'):
                info += ','.join([x_min,y_min,x_max,y_max])
                info = ','.join([info,'1'])
                cnt += 1
            
            else:
                pass   
                

        elif ((len(value[0]) != 0) & (cnt>0)):
            
            x_min = str(value[0]['X좌표'])
            x_max = str(value[0]['X좌표']+value[0]['가로'])
            y_min = str(value[0]['Y좌표'])
            y_max = str(value[0]['Y좌표']+value[0]['세로'])
            
            if (cate=='하의'):
                info += ' ' 
                info += ','.join([x_min,y_min,x_max,y_max])
                info = ','.join([info,'0'])
                cnt += 1
                
            elif (cate=='상의'):
                info += ' '
                info += ','.join([x_min,y_min,x_max,y_max])
                info = ','.join([info,'1'])
                cnt += 1
           
            else:
                pass

    print(path)
    print(info)
    
    # 레이블이 존재하지 않는 경우 -> 추가하지 않음
    if len(info) < 5:
        pass
    else:    
        # 하나의 txt 파일 저장
        storage_form = r'D:\K-Fashion 이미지\label_my\feminine.txt' ###영어로 경로 지정하기!
        f = open(storage_form,'a',encoding='utf-8')
        f.write('\n')
        f.write(path)
        f.write(' ')
        f.write(info)
        f.close()
```
